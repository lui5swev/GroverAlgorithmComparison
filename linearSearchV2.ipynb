```python
import csv
import time
import random


csvName = "database1m.csv"

database = []
with open(csvName, newline="") as f:
    reader = csv.DictReader(f)
    for row in reader:
        database.append(int(row["valor"]))

print(f"Registros cargados: {len(database)}")

# Elegir un valor aleatorio del archivo
target = random.choice(database)
print(f"Elemento a buscar: {target}")
```

```python
def linear_search(data, target):
    for i, val in enumerate(data):
        if val == target:
            return i
    return -1

start = time.perf_counter()
index = linear_search(database, target)
end = time.perf_counter()

print(f"Resultado clásico -> Índice: {index}, Valor: {database[index]}")
print(f"Tiempo de búsqueda lineal: {end - start:.6f} segundos")
```
